### Problem 1

Use three variables 'Height', 'Length' and 'Rings' from abalone dataset to make a scatter matrix like this:
<img src=http://slides.nycdatascience.com/BOOTCAMP3/content/week6/pic/scatter_matrix.png>

```python
import pandas as pd
import matplotlib as plt
abalone = pd.read_csv('./abalone.data', header = 0,
                      names = ['Sex', 'Length', 'Diam', 'Height', 'Whole', 'Shucked', 'Viscera', 'Shell', 'Rings'])
```

```python
abalone.head()
```

```python
Height = abalone.Height
Length = abalone.Length
Rings = abalone.Rings
```

- Figure size: width 12; height 12.

- Transparency: 0.5

- Pay attention to the labels on the axes.

```python
%matplotlib inline
df = pd.DataFrame({'Height': Height, 'Length': Length, 'Rings': Rings })
pd.scatter_matrix(df, alpha=.1, figsize=(10, 10), diagonal='hist')
```

```python
google = pd.read_csv('./google.csv', index_col=0)
google = google.drop('Volume', axis = 1)
google.head()
```

```python
### write your codes here
import datetime as dt
import pylab as plt

### write your codes here
highs=google.iloc[:,1][::-1]
lows=google.iloc[:,2][::-1]

plt.figure(figsize=(12,5))
plt.plot(np.arange(len(highs)), highs, color = 'red', label='High')
plt.plot(np.arange(len(lows)), lows, color = 'gray', label='Low')

plt.legend(loc=1)
plt.title('Daily prices of Google', fontsize = 25)
plt.xlabel('Date', fontsize = 15)
```

![density](http://slides.nycdatascience.com/BOOTCAMP3/content/week6/pic/exercise_density.png)

```python
mtcars = pd.read_csv("./mtcars.csv")
mtcars.head()
```

```python
### write your codes here
import seaborn as sns

three = mtcars[mtcars['gear']==3]
four = mtcars[mtcars['gear']==4]
five = mtcars[mtcars['gear']==5]


plt.figure(figsize=(12,12))

sns.kdeplot(three['mpg'], shade=True, color="blue",label='Third Gear',alpha=0.3, legend = False)
sns.kdeplot(four['mpg'], shade=True, color="green",label='Fourth Gear',alpha=0.3, legend = False)
sns.kdeplot(five['mpg'], shade=True, color="red",label='Fifth Gear',alpha=0.3, legend = False)

plt.legend(loc='center right')
plt.ylabel('Density')
plt.title('Distribution of Gas Milage')
plt.xlabel('Miles per Gallon (mpg)')

plt.show()
```
